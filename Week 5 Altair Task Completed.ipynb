```python
import altair as alt
import pandas as pd

my_file = pd.read_csv("2020 Constructors Championship.csv")
my_file
```

```python
brush = alt.selection_interval()

points = alt.Chart(my_file).mark_bar().encode(
    x='Year:N',
    y='Position:N',
    color=alt.condition(brush, 'Team:N', alt.value('gray'))
).add_selection(
    brush
)
points
```

```python
bars = alt.Chart(my_file).mark_line().encode(
    x='Year:N',
    y='Position:N',
    color=alt.condition(brush, 'Team:N', alt.value('gray'))
).add_selection(
    brush
)
bars
```

I have chosen the below designs as I wanted the data to be easy to read and interpret. The data is the Constructors Championship Standings from 2018 - 2020.  The data shows the improvement / decline of all teams across a 3 year period. Mercedes clearly dominate across the 3 years, Ferrari and Haas show the biggest decline dropping 4 places and Toro Rosso & Racing Point show the biggest improvement gaining 3 positions.
